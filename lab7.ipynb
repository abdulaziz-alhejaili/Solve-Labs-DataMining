```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

```python
from sklearn import datasets
iris= datasets.load_iris()
df= pd.read_csv(iris.filename)
```

```python
x = df.iloc[:, [0,1,2,3]].values
```

# Exercise 1.1
<p> K Means algorithm is an iterative algorithm that tries to partition the dataset
into K Predefined distinct non-overlapping subgroups (clusters)
where each data point belongs to only one group
minimum parameters which are necessary to execute it is 0
because all the parameters have a default value that will assigned to execute it <p> 

```python
kmeans5 = KMeans(n_clusters=5)
y_kmeans5 = kmeans5.fit_predict(x)
print(y_kmeans5)
kmeans5.cluster_centers_
```

# Repeat the experiment again and set the k as 2. Compare the outputs

```python
kmeans2 = KMeans(n_clusters=2)
y_kmeans2 = kmeans2.fit_predict(x)
print(y_kmeans2)
kmeans2.cluster_centers_
```

# First EX take numbers from 0 to 4 equals 5 numbers and have 5 rows.

# Second EX t take numbers from 0 to 1 equals 2 numbers and have 2 rows

```python
Error =[]
for i in range(1,11):
    kmeans = KMeans(n_clusters = i).fit(x)
    kmeans.fit(x)
    Error.append(kmeans.inertia_)
plt.plot(range(1, 11), Error)
plt.title('Elbow methpd')
plt.xlabel('No of clusters')
plt.ylabel('Error')
plt.show()
```
