# 3.2.1 Scatter Plots

```python
df.plot(kind ='scatter',x='sepal length (cm)',y='sepal width (cm)',c='g')
plt.show()
```

```python
sns.set_style("whitegrid");
sns.FacetGrid(df, hue="class",height=10)\
   .map(plt.scatter,"sepal length (cm)" ,"sepal width (cm)")\
   .add_legend();
plt.show();
```

# 3.2.2 Pair Plots

```python
sns.set_style("whitegrid");
sns.pairplot(df, vars = iris.feature_names, hue ='class', diag_kind = 'kde',
            plot_kws = {'alpha':0.6, 's':80, 'edgecolor': 'k'}, size = 2)
plt.show()
```
